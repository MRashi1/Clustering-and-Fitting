```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score, silhouette_samples
from scipy.stats import skew, kurtosis

# Load the dataset
iris = pd.read_csv('iris.csv')

# Display the first few rows of the dataset
print("First few rows of the dataset:")
print(iris.head())

def calculate_descriptive_statistics(data):
    # Calculate descriptive statistics
    statistics = data.describe()
    
    # Calculate skewness and kurtosis for numerical columns
    numeric_cols = data.select_dtypes(include=[np.number])
    skewness = numeric_cols.apply(skew)
    kurt = numeric_cols.apply(kurtosis)
    
    # Add skewness and kurtosis to statistics
    statistics.loc['Skewness'] = skewness
    statistics.loc['Kurtosis'] = kurt
    
    return statistics

def filter_descriptive_statistics(statistics):
    required_columns = ['mean', '50%', 'std', 'Skewness', 'Kurtosis']
    return statistics.loc[required_columns].rename(index={'mean': 'Mean', '50%': 'Median', 'std': 'Standard Deviation', 'Skewness': 'Skewness', 'Kurtosis': 'Kurtosis'})

def print_descriptive_statistics(statistics):
    print("\nDescriptive Statistics:")
    print(statistics)

# Calculate descriptive statistics
descriptive_stats = calculate_descriptive_statistics(iris)

# Filter and print descriptive statistics
filtered_stats = filter_descriptive_statistics(descriptive_stats)
print_descriptive_statistics(filtered_stats)

def plot_histogram(data, feature):
    plt.figure(figsize=(8, 6))
    plt.hist(data[feature], bins=20, edgecolor='black')
    plt.title(f'Histogram of {feature}')
    plt.xlabel(feature)
    plt.ylabel('Frequency')
    plt.show()

def plot_scatter(data, x, y, hue):
    plt.figure(figsize=(8, 6))
    plt.scatter(data[x], data[y], c=data[hue].map({'Iris-setosa': 0, 'Iris-versicolor': 1, 'Iris-virginica': 2}))
    plt.title(f'Scatter Plot of {x} vs {y}')
    plt.xlabel(x)
    plt.ylabel(y)
    plt.colorbar(ticks=[0, 1, 2], label=hue)
    plt.show()

def plot_violin(data, x, y, hue):
    plt.figure(figsize=(8, 6))
    plt.violinplot(dataset=[data[y][data[hue] == 'Iris-setosa'],
                            data[y][data[hue] == 'Iris-versicolor'],
                            data[y][data[hue] == 'Iris-virginica']],
                   showmeans=True)
    plt.xticks([1, 2, 3], ['Iris-setosa', 'Iris-versicolor', 'Iris-virginica'])
    plt.title(f'Violin Plot of {y} by {hue}')
    plt.ylabel(y)
    plt.show()

def plot_silhouette(data, n_clusters):
    X = data.drop(['Species'], axis=1)
    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    kmeans.fit(X)
    silhouette_avg = silhouette_score(X, kmeans.labels_)
    print(f"Silhouette Score: {silhouette_avg:.3f}")

    plt.figure(figsize=(8, 6))
    sample_silhouette_values = silhouette_samples(X, kmeans.labels_)
    plt.hist(sample_silhouette_values, bins=20, edgecolor='black')
    plt.axvline(x=silhouette_avg, color='red', linestyle='--')
    plt.title('Silhouette Plot')
    plt.xlabel('Silhouette Coefficient')
    plt.ylabel('Frequency')
    plt.show()

def plot_line_fit(data, x, y):
    plt.figure(figsize=(8, 6))
    species_means = data.groupby('Species')[x].mean()
    species_means.plot(marker='o')
    plt.title(f'Line Plot of {x} vs {y}')
    plt.xlabel('Species')
    plt.ylabel(x)
    plt.xticks(rotation=45)
    plt.grid(True)
    plt.show()

if __name__ == '__main__':
    # Plot a histogram/bar chart/pie chart
    plot_histogram(iris, 'PetalLengthCm')

    # Plot a line/scatter graph
    plot_scatter(iris, 'PetalLengthCm', 'PetalWidthCm', 'Species')

    # Plot a confusion matrix/heatmap/corner/box/violin plot
    plot_violin(iris, 'Species', 'SepalWidthCm', 'Species')

    # Plot an elbow/silhouette plot
    plot_silhouette(iris, n_clusters=3)

    # Plot line fitting
    plot_line_fit(iris, 'PetalLengthCm', 'Species')
```
